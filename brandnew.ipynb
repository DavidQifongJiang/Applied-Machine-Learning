```python
import re
import nltk
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import f1_score, accuracy_score
from scipy.sparse import vstack
from scipy.optimize import linear_sum_assignment
from nltk.corpus import stopwords, wordnet as wn
from nltk.stem import WordNetLemmatizer, PorterStemmer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Initialize stemmer and lemmatizer
lemmer = WordNetLemmatizer()
stemmer = PorterStemmer()

# Tag dictionary for POS mapping
TAG_DICT = {
    'J': wn.ADJ,
    'V': wn.VERB,
    'N': wn.NOUN,
    'R': wn.ADV
}

# Define cleaning and preprocessing functions
def clean(text):
    text = re.sub(r'https?://\S+|www\.\S+', '', str(text))  # Remove URLs
    text = re.sub(r"<br />", " ", text)
    text = re.sub(r"&quot;", "\"", text)
    text = re.sub('&#39;', "\"", text)
    text = re.sub('\n', " ", text)
    text = re.sub(' u ', " you ", text)
    text = re.sub('`', "", text)
    text = re.sub(r"(!)\1+", r"!", text)
    text = re.sub(r"(\?)\1+", r"?", text)
    text = re.sub('&amp;', 'and', text)
    text = re.sub(r'\W', ' ', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text)  # Substitute multiple spaces with a single space
    text = text.lower().strip()  # Convert to lowercase and strip whitespace
    return text

def clean_dataset(dataset):
    dataset['Phrase'] = dataset['Phrase'].apply(clean)
    return dataset

def tokenize_lexicon(texts):
    return [nltk.pos_tag(nltk.word_tokenize(text)) for text in texts]

def get_wordnet_pos(pos_tag):
    return TAG_DICT.get(pos_tag[0], wn.NOUN)

def lemmatize_texts(texts):
    return [
        [lemmer.lemmatize(word, pos=get_wordnet_pos(pos_tag)) for word, pos_tag in text]
        for text in texts
    ]

def stem_texts(texts):
    """
    Stem each word in the tokenized texts.
    """
    return [
        [stemmer.stem(word) for word in text]  # Stem each word without needing POS tags
        for text in texts
    ]

def backtostring(texts):
    return [" ".join(text) for text in texts]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import IncrementalPCA
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from scipy.sparse import vstack
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
from scipy.optimize import linear_sum_assignment
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression

# Vectorize phrases (1 for Bag of Words, 2 for N-gram)
def vectorize_text(vectorization_type, training_set):
    if vectorization_type == 1:
        text_vectorizer = CountVectorizer(binary=True, min_df=6)
    elif vectorization_type == 2:
        text_vectorizer = CountVectorizer(binary=True, min_df=2, ngram_range=(2, 2))
    else:
        raise ValueError("Invalid vectorization type specified.")
    text_vectorizer.fit(training_set['Phrase'])
    return text_vectorizer

# Unsupervised learning with IncrementalPCA and clustering
def unsupervised_clustering(cluster_method, num_clusters, X_unlabeled, X_labeled, y_labels):
    # Use Incremental PCA to handle large data
    pca_model = IncrementalPCA(n_components=100, batch_size=500)  # Adjusted batch size for memory efficiency
    X_unlabeled_reduced = pca_model.fit_transform(X_unlabeled)
    X_labeled_reduced = pca_model.transform(X_labeled)

    # Select the clustering model
    if cluster_method == 1:
        cluster_model = KMeans(n_clusters=num_clusters, random_state=42)
    elif cluster_method == 2:
        cluster_model = GaussianMixture(n_components=num_clusters, covariance_type='diag', random_state=42, max_iter=100, tol=1e-3)
    else:
        raise ValueError("Invalid clustering method. Use 1 for KMeans or 2 for GaussianMixture.")

    # Generate pseudo-labels
    pseudo_labels_unlabeled = cluster_model.fit_predict(X_unlabeled_reduced)

    # Map clusters to labels for best alignment
    alignment_matrix = np.zeros((num_clusters, num_clusters))
    for cluster_idx in range(num_clusters):
        labeled_samples = y_labels[cluster_model.predict(X_labeled_reduced) == cluster_idx]
        for label in range(num_clusters):
            alignment_matrix[cluster_idx, label] = np.sum(labeled_samples != label)

    # Apply Hungarian algorithm for cluster alignment
    row_indices, col_indices = linear_sum_assignment(alignment_matrix)
    cluster_to_label_mapping = {row: col for row, col in zip(row_indices, col_indices)}
    pseudo_labels_aligned = np.array([cluster_to_label_mapping[cluster] for cluster in pseudo_labels_unlabeled])

    # Combine labeled and pseudo-labeled data without converting to dense format
    X_full = vstack([X_labeled, X_unlabeled])
    y_full = np.concatenate((y_labels, pseudo_labels_aligned))

    return X_full, y_full

# Supervised learning (1 for Multinomial Naive Bayes, 2 for Softmax Regression)
def supervised_learning(training_method, X_data, y_data, X_validation, y_validation):
    if training_method == 1:
        model = MultinomialNB(alpha=1) 
    elif training_method == 2:
        model = LogisticRegression(solver='lbfgs', max_iter=1000)  # 'multinomial' is default for multiclass
    else:
        raise ValueError("Invalid supervised learning method specified.")

    model.fit(X_data, y_data)

    # Predict and evaluate on validation set
    predictions_val = model.predict(X_validation)
    accuracy_val = accuracy_score(y_validation, predictions_val)
    f1_val = f1_score(y_validation, predictions_val, average="macro")
    
    return training_method, accuracy_val, f1_val

# Main training pipeline
def train_pipeline(preprocess_type, cluster_type, model_type):
    # Step 1: Vectorize the text data
    vectorizer = vectorize_text(preprocess_type, train_data)
    
    X_labeled_train = vectorizer.transform(labeled_train['Phrase'])  # Kept as sparse format
    X_labeled_val = vectorizer.transform(labeled_val['Phrase'])
    X_unlabeled_train = vectorizer.transform(unlabeled_train['Phrase'])
    
    y_labeled_train = labeled_train['Sentiment']
    y_val_set = labeled_val['Sentiment']
    num_classes = len(np.unique(y_labeled_train))  # Calculate number of classes

    # Step 2: Apply unsupervised learning to create pseudo-labels
    X_combined_data, y_combined_labels = unsupervised_clustering(cluster_type, num_classes, X_unlabeled_train, X_labeled_train, y_labeled_train)
    
    # Step 3: Train the supervised model on the combined dataset
    model, validation_accuracy, validation_f1 = supervised_learning(model_type, X_combined_data, y_combined_labels, X_labeled_val, y_val_set)
    
    # Output performance metrics
    print(f"Validation Accuracy: {validation_accuracy:.4f}")
    print(f"F1 Score on Validation Set: {validation_f1:.4f}")

# Running configurations for the training pipeline
for preproc_type in range(1, 3):
    for cluster_type in range(1, 3):
        for model_type in range(1, 3):
            # Define the options for preprocessing, clustering, and supervised models
            preprocessing_options = ["Bag of Words", "N-gram"]
            clustering_options = ["K-Means", "Gaussian Mixture"]
            model_options = ["Multinomial Naive Bayes", "Softmax Regression"]
            
            # Create a descriptive model configuration string
            config_description = f'\nUsing: {preprocessing_options[preproc_type - 1]}, {clustering_options[cluster_type - 1]}, {model_options[model_type - 1]}'
            print(config_description)
            
            # Run the final training pipeline
            train_pipeline(preproc_type, cluster_type, model_type)
            print("\n")
```
